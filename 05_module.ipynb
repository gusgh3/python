## random

````python
import random
````

````python
random.random() #0~1까지의 소수 중 랜덤으로 출력
````

````python
random.randint(1, 10) #1~10까지의 숫자 중 랜덤으로 출력
````

````python
random.seed(5)  #seed() 괄호안에는 어떤 숫자가 들어가도 상관 없음. seed함수 자체가 도출되는 값을 고정시킨다는 의미로 보면 됨
random.random()
````

````python
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers) #함수 내 인자(대이터)들이 랜덤으로 섞인 값이 계속 나옴
````

````python
random.choice(numbers) #함수 내 인자들이 랜덤으로 나옴
````

````python
print(random.choice(numbers))
print(random.choice(numbers))
````

## 구현해야 할 내용

-requests를 통해서 이번주 당첨번호를 가져오고 lsit에 넣어 출력하기
-5회차 번호를 sample 함수를 이용해 번호 6개 출력
-정답 list랑 dict안에 있는 회차 번호랑 비교해서 1-5등, 꽝여부 비교

````python
import random
````

````python
numbers = list(range(1,46))
print(numbers)
````

````python
random.sample(numbers, 6)
````

````python
buy_dict = {}
for i in range(1,6):
    buy_dict[f'{i}회차'] = random.sample(numbers, 6)

print(buy_dict)
````

````python
#pair programming 답안

payload = {
    'method': 'getLottoNumber',
    'drwNo': 1126,
}

r = requests.get('https://www.dhlottery.co.kr/common.do?method=getLottoNumber&drwNo=1126')
# print(r.json())
# print(type(r.json()))

lotto_dict = r.json()

numbers = list(range(1,46))
buy_dict = {}
for i in range(1,6):
    buy_dict[f'{i}회차'] = random.sample(numbers, 6)


correct_list = []
bonus = lotto_dict['bnusNo']
for i in range(1,7):
    correct_list.append(lotto_dict[f'drwtNo{i}'])
print(f'이번주 당첨번호 : {correct_list}, BN: {bonus}')

print(buy_dict)

for key, value in buy_dict.items():
    cnt = 0
    for i in value:
        if i in correct_list:
            cnt += 1
    if cnt == 6:
        print("1등")
    elif cnt == 5:
        if bonus in value:
            print("2등")
        else:
            print("3등")
    elif cnt == 4:
        print("4등")
    elif cnt == 3:
        print("5등")
    else:
        print("꽝")
````
